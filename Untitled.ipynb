```python
def boxp(t):
    f,a=mp.subplots(t.shape[1],figsize=(20,15))
    for i,j in enumerate(t.columns):
        f.tight_layout(pad=0.1)
        a[i].boxplot(t[j],vert=0);
        a[i].set_xlabel(j,size=25)
```

```python
def distp(t):
    fi,ax=mp.subplots(t.shape[1],figsize=(20,15))
    for i,j in enumerate(t.columns):
        fi.tight_layout(pad=0.1)
        sn.distplot(t[j],ax=ax[i])
        ax[i].set_xlabel(j,size=25)
```
